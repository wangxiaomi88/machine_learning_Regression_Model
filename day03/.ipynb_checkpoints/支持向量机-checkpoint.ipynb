```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn.datasets as sd
import sklearn.preprocessing as sp
import sklearn.linear_model as lm
import sklearn.tree as st
import sklearn.ensemble as se
import sklearn.svm as svm
import sklearn.model_selection as ms
import sklearn.metrics as sm

```

```python
data = pd.read_csv("multiple2.txt",header=None,names=["px","py","target"])
data
```

```python
plt.scatter(data["px"],data["py"],c=data["target"],cmap="jet",s=60,alpha=0.5)
plt.colorbar()
```

```python
# æ‹†åˆ†è®­ç»ƒé›†ã€æµ‹è¯•é›†

x = data.iloc[:,:-1] #å–åŽ100ä¸ªæ•°æ®ï¼Œåªæ¶‰åŠäºŒåˆ†ç±»
y = data.iloc[:,-1] #å–åŽ100ä¸ªæ•°æ®ï¼Œåªæ¶‰åŠäºŒåˆ†ç±»

train_x,test_x,train_y,test_y = ms.train_test_split(x,y,test_size=0.2,random_state=7,stratify=y)

model_svm = svm.SVC(kernel="linear")
model_svm.fit(train_x,train_y)

pred_test_y = model_svm.predict(test_x)

m = sm.confusion_matrix(test_y,pred_test_y)
print(m)
cr = sm.classification_report(test_y,pred_test_y)
print(cr)
```

```python
# æš´åŠ›ç»˜åˆ¶åˆ†ç±»è¾¹ç•Œçº¿

#ä»Žxçš„minåˆ°maxæ‹†å‡º100ä¸ªåæ ‡
#ä»Žyçš„minåˆ°maxæ‹†å‡º100ä¸ªåæ ‡

# ä¸€å…±ç»„æˆ10000ä¸ªç‚¹ï¼Œé¢„æµ‹æ¯ä¸ªæ•£ç‚¹çš„ç±»åˆ«ï¼Œç»˜åˆ¶æ•£ç‚¹å›¾

xs=np.linspace(data["px"].min(),data["px"].max(),100)
ys=np.linspace(data["py"].min(),data["py"].max(),100)


xx,yy = np.meshgrid(xs,ys)


xx = xx.reshape(-1,)
yy = yy.reshape(-1,)

xx.shape

points=pd.DataFrame([])

points["px"]=pd.Series(xx)
points["py"]=pd.Series(yy)

points["target"] = model_svm.predict(points)

plt.scatter(points["px"],points["py"],c=points["target"],cmap="gray",s=10,alpha=0.2)
plt.scatter(data["px"],data["py"],c=data["target"],cmap="jet",s=60,alpha=0.8)
plt.colorbar()





```

 # é€šè¿‡ç½‘æ ¼æœç´¢ï¼Œå¯»æ±‚æœ€ä¼˜è¶…å‚æ•°ç»„åˆ

```python
model_base = svm.SVC()


#ç½‘æ ¼æœç´¢
params = [{"kernel":["linear"],"C":[1,10,100]},
          {"kernel":["poly"],"C":[2,3]},
          {"kernel":["rbf"],"C":[1,10,100],"gamma":[1,0.1,0.01]}]

model_svm = ms.GridSearchCV(model_base,params,cv=5)


# æ‹†åˆ†è®­ç»ƒé›†ã€æµ‹è¯•é›†
x = data.iloc[:,:-1] #å–åŽ100ä¸ªæ•°æ®ï¼Œåªæ¶‰åŠäºŒåˆ†ç±»
y = data.iloc[:,-1] #å–åŽ100ä¸ªæ•°æ®ï¼Œåªæ¶‰åŠäºŒåˆ†ç±»
train_x,test_x,train_y,test_y = ms.train_test_split(x,y,test_size=0.25,random_state=7,stratify=y)



# æ¨¡åž‹è®­ç»ƒ
model_svm.fit(train_x,train_y)

pred_test_y = model_svm.predict(test_x)

m = sm.confusion_matrix(test_y,pred_test_y)
print(m)
cr = sm.classification_report(test_y,pred_test_y)
print(cr)

# æš´åŠ›ç»˜åˆ¶åˆ†ç±»è¾¹ç•Œçº¿

#ä»Žxçš„minåˆ°maxæ‹†å‡º100ä¸ªåæ ‡
#ä»Žyçš„minåˆ°maxæ‹†å‡º100ä¸ªåæ ‡

# ä¸€å…±ç»„æˆ10000ä¸ªç‚¹ï¼Œé¢„æµ‹æ¯ä¸ªæ•£ç‚¹çš„ç±»åˆ«ï¼Œç»˜åˆ¶æ•£ç‚¹å›¾

xs=np.linspace(data["px"].min(),data["px"].max(),100)
ys=np.linspace(data["py"].min(),data["py"].max(),100)


xx,yy = np.meshgrid(xs,ys)


xx = xx.reshape(-1,)
yy = yy.reshape(-1,)

xx.shape

points=pd.DataFrame([])

points["px"]=pd.Series(xx)
points["py"]=pd.Series(yy)

points["target"] = model_svm.predict(points)

plt.scatter(points["px"],points["py"],c=points["target"],cmap="gray",s=10,alpha=0.2)
plt.scatter(data["px"],data["py"],c=data["target"],cmap="jet",s=60,alpha=0.8)
plt.colorbar()





print(model_svm.best_params_)
print(model_svm.best_score_)
print(model_svm.best_estimator_)



```
